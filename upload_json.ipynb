```python
import warnings
import httpx
import orjson
import os
import argparse

warnings.filterwarnings("ignore")  # ÂøΩÁï• SSL Ë≠¶Âëä

def upload_json_aiteam(json_file):                               
    UPLOAD_DATA_JSON_URL = 'http://localhost:84/api/ai_team/create_or_update'
    client = httpx.Client(timeout=10.0, verify=False)  # Âä†ÂÖ• timeout/‰∏çÈ©óË≠â SSL

    def post_json(path):
        with open(path, 'rb') as f:
            data = orjson.loads(f.read())
        response = client.post(UPLOAD_DATA_JSON_URL, json=data)
        print(f"\nüì§ Uploaded: {os.path.basename(path)}")
        print(f"‚úÖ Status: {response.status_code}")
        try:
            print("üì• Response:", response.json())
        except Exception:
            print("üì• Response (raw):", response.text)

    with client:
        if isinstance(json_file, str):
            post_json(json_file)
        elif isinstance(json_file, list):
            for path in json_file:
                post_json(path)
        else:
            print("‚ùå Invalid input: must be a str or list of str")
```

```python
# Example usage
json_file = '/data/4TB1/pipeline/chuan/process/Deep_Aneurysm/07807990_20250715_MR_21405290051/nnUNet/JSON/07807990_20250715_MR_21405290051_platform_json.json'
upload_json_aiteam(json_file)
```
